### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data 

```python
driver.get("https://www.naukri.com/")
```

```python
## Now finding the elecment from website 
job_search =driver.find_element_by_id("qsb-keyword-sugg")
job_search.send_keys("Data Analyst")
job_location =driver.find_element_by_xpath("//input[@id='qsb-location-sugg']")
job_location.send_keys("Bangalore")
click_bt=driver.find_element_by_xpath("//div[@class='search-btn']/button")
click_bt.click()
```

```python
## Extracting the detail from the website
job_title=[]
job_loc=[]
comp_name=[]
Exp=[]
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    job_title.append(i.text)
job_title= job_title[0:10]
for i in driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span"):
    job_loc.append(i.text)
job_loc= job_loc[0:10]
for i in driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']"):
    comp_name.append(i.text)
comp_name= comp_name[0:10]
for i in driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span"):
    Exp.append(i.text)
Exp= Exp[0:10]
df= pd.DataFrame({"Job Title":job_title,"Location":job_loc, "Company":comp_name,"Experience":Exp})
df
```
