### Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description

```python
## entering the website to the browser
driver.get("https://www.hostelworld.com/") 
```

```python
search=driver.find_element_by_xpath("//input[@class = 'search-input']")
search.send_keys("London")
```

```python
driver.find_element_by_xpath("//div[@class='label']").click()
time.sleep(2)
driver.find_element_by_id("search-button").click()
```

```python
Name =[]
dis =[]
rat=[]
rev=[]
pri_price=[]
dor_price=[]
fac=[]
desc=[]
for i in driver.find_elements_by_xpath("//h2[@class='title title-6']/a"):
    Name.append(i.text)
for i in driver.find_elements_by_xpath("//span[@class='description']"):
    dis.append(i.text)
for i in driver.find_elements_by_xpath("//div[@class='score orange big']"):
    rat.append(i.text)
for i in driver.find_elements_by_xpath("//div[@class='reviews']"):
    rev.append(i.text)
for i in driver.find_elements_by_xpath("//div[@class='price title-5']"):
    pri_price.append(i.text)
pri_p=pri_price[1:52:2]
dor_price=pri_price[0:52:2]
for i in driver.find_elements_by_xpath("//div[@class='facilities-label facilities']"):
     fac.append(i.text)
for i in driver.find_elements_by_xpath("//div[@class='rating-factors prop-card-tablet rating-factors small']"):
     desc.append(i.text)
        
rat.append(0)
pri_p = pri_p+[0,0,0,0]
dor_price= dor_price+[0,0,0,0]

print(len(Name),len(dis),len(rat),len(rev),len(pri_p),len(dor_price),len(fac),len(desc))
df= pd.DataFrame({'Hostel Name': Name, 'Distance from city centre':dis, 'Ratings':rat, 'Total reviews':rev, 
 'Privates from price':pri_p, 'Dorms from price':dor_price, 'Facilities and property':fac , 'Description':desc})
df
```
