# Q-6 Posts from coreyms.com

```python
import re
page = requests.get("https://coreyms.com/")
page.content

soup= BeautifulSoup(page.content)
Header = []
for i in soup.find_all("h2",class_="entry-title"):
    Header.append(i.get_text())
date=[]
for x in soup.find_all("time", class_="entry-time"):
    date.append(x.get_text())
content=[]
for x in soup.find_all("div",class_="entry-content"):
    content.append(x.get_text().replace("\n"," "))
link=[]
for x in soup.find_all('a', attrs={'href': re.compile("^https://coreyms.com/development/python")}):
    link.append(x.get("href"))
link = link[0:19:2]
Header.remove("Update (2019-09-03)")
len(Header)
date.remove("September 3, 2019")
del content[4]
df = pd.DataFrame({"Heading": Header, "Date":date, "Content":content, "Link": link})
df
```
