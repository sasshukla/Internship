### Scrape the details of most viewed videos on YouTube from Wikipedia.
### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

```python
### Iniliazing the chrome
driver = webdriver.Chrome(r'E:\Download backup\chromedriver_win32\chromedriver.exe') 
```

```python
## entering the website to the browser
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos") 
```

```python
rank = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[1]")
vname = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[2]")
aname=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[3]")
udate= driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[4]")
view = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[5]")
```

```python
top_video=[]
for i in range(len(rank)):
    data ={'Rank':rank[i].text,
          'Video Name':vname[i].text,
          'Artist Name':aname[i].text,
          'Upload Date':udate[i].text,
          'Views on video':view[i].text,}
    top_video.append(data)
df= pd.DataFrame(top_video)
df
```

### Scrape the details most watched tv series of all time from imdb.com.

```python
name =[]
Year=[]
Genre=[]
Run_time=[]
Ratings=[]
Votes=[]
for i in range(1,101):
    for j in driver.find_elements_by_xpath("//*[@id='main']/div/div[3]/div[3]/div"+str([i])+"/div[2]/h3/a"):
        name.append(j.text)
    for j in driver.find_elements_by_xpath("//*[@id='main']/div/div[3]/div[3]/div"+str([i])+"/div[2]/h3/span[2]"):
        Year.append(j.text)
    for j in driver.find_elements_by_xpath("//*[@id='main']/div/div[3]/div[3]/div"+str([i])+"/div[2]/p[1]/span[5]"):
        Genre.append(j.text)
    for j in driver.find_elements_by_xpath("//*[@id='main']/div/div[3]/div[3]/div"+str([i])+"/div[2]/p[1]/span[3]"):
        Run_time.append(j.text)
    for j in driver.find_elements_by_xpath("//*[@id='main']/div/div[3]/div[3]/div"+str([i])+"/div[2]/div[1]/div[1]/span[2]"):
        Ratings.append(j.text)
    for j in driver.find_elements_by_xpath("//*[@id='main']/div/div[3]/div[3]/div"+str([i])+"/div[2]/p[4]/span[2]"):
        Votes.append(j.text)
df=pd.DataFrame({"Name":name,"Year Span":Year,"Genre":Genre,"Run Time":Run_time,"Ratings":Ratings,"Votes":Votes})
df
```

### Details of Datasets from UCI machine learning repositories.
### Url = https://archive.ics.uci.edu/

```python
driver = webdriver.Chrome(r'E:\Download backup\chromedriver_win32\chromedriver.exe') 
driver.get("https://archive.ics.uci.edu/")
```

```python
driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a").click()
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
instances=[]
attribute=[]
year=[]
for i in range(2,623):
    for i in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr"+str([i])+"/td[1]/table/tbody/tr/td[2]"):
        Dataset_name.append(i.text)
    
```

```python
for i in range(2,623):
    for j in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr"+str([i])+"/td[2]/p"):
        Data_type.append(j.text)
    for j in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr"+str([i])+"/td[3]/p"):
        Task.append(j.text)
    for j in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr"+str([i])+"/td[4]/p"):
        Attribute_type.append(j.text)
    for j in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr"+str([i])+"/td[5]/p"):
        instances.append(j.text)
    for j in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr"+str([i])+"/td[6]/p"):
        attribute.append(j.text)
    for j in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr"+str([i])+"/td[7]/p"):
        year.append(j.text)
print(len(Dataset_name),len(Data_type),len(Task),len(Attribute_type),len(instances),len(attribute),len(year))
```

```python
df=pd.DataFrame({"Dataset Name":Dataset_name, "Data Type": Data_type, "Task":Task,"Attribute_type":Attribute_type,"Instances":instances, "Attribute":attribute, "Year":year})
df
```
